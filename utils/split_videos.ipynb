```python
import os
import cv2

VIDEO_DIR = "/data/ephemeral/home/videos_movieclips"

video_files = [os.path.join(VIDEO_DIR, f) for f in os.listdir(VIDEO_DIR) if f.endswith(".mp4")]

corrupted_files = []

for video_path in video_files:
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"손상된 파일 : {video_path}")
        corrupted_files.append(video_path)
    cap.release()
    
if corrupted_files:
    with open("corrupted_videos.txt", "w") as f:
        for file in corrupted_files:
            f.write(file + "\n")
    print(f"🚨 총 {len(corrupted_files)}개의 손상된 파일이 있습니다. `corrupted_videos.txt` 파일에 저장됨.")
else:
    print("✅ 모든 비디오 파일이 정상적으로 열립니다!")
```

```python
total_videos = len(video_files)

half = total_videos // 2

server1_videos = video_files[:half]
server2_videos = video_files[half:]

with open("server1_videos.txt", "w") as f:
    for file in server1_videos:
        f.write(file + "\n")

with open("server2_videos.txt", "w") as f:
    for file in server2_videos:
        f.write(file + "\n")

print(f"총 {total_videos}개의 비디오 중 {len(server1_videos)}개는 Server 1, {len(server2_videos)}개는 Server 2에서 사용됩니다.")
print("`server1_videos.txt`와 `server2_videos.txt`에 저장됨.")
```

```python
import os
import shutil

VIDEO_DIR = "/data/ephemeral/home/videos_movieclips"

TXT_FILE = "/data/ephemeral/home/server2_videos.txt"

ANOTHER_DIR = os.path.join(VIDEO_DIR, "another")

if not os.path.exists(ANOTHER_DIR):
    os.makedirs(ANOTHER_DIR)

with open(TXT_FILE, "r") as f:
    keep_files = set(os.path.basename(line.strip()) for line in f.readlines())

print(f"📄 유지할 파일 개수: {len(keep_files)}")
print(f"📂 비디오 폴더 내 파일 개수: {len(os.listdir(VIDEO_DIR))}")

for file_name in os.listdir(VIDEO_DIR):
    file_path = os.path.join(VIDEO_DIR, file_name)

    if file_name not in keep_files and os.path.isfile(file_path):
        shutil.move(file_path, os.path.join(ANOTHER_DIR, file_name))
        print(f"🚚 Moved: {file_name} → {ANOTHER_DIR}")

print("✅ 정리 완료! 필요한 파일만 남기고, 나머지는 `another/`로 이동했습니다.")
```
